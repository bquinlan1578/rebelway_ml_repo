```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv("/Users/felipepesantez/Documents/development/python/rway_dev/repos/ml_course/machine_learning/knn/factions_data.csv")
```

```python
df.head()
```

```python
df.drop(["Object_class","Position Y", "Point"],inplace=True, axis=1)
```

```python
df.head()
```

```python
def g_distance(x):
    res = 0
    if x == "low":
        res  = 0/2
    if x == "medium":
        res = 1/2
    if x == "high":
        res = 2/2

    return res
```

```python
df['social_class'] = df['social_class'].apply(g_distance)
```

```python
df.head()
```

```python
scaler = StandardScaler()
scaler.fit(df.drop(["faction", "social_class"],axis=1))
```

```python
scaled_features = scaler.transform(df.drop(['faction','social_class'],axis=1))
```

```python
scaled_features
```

```python
df_scaled_features = pd.DataFrame(scaled_features, columns=df.drop(['faction', 'social_class'],axis=1).columns)
```

```python
df_scaled_features.head()
```

```python
df_scaled_features['social_class'] = df['social_class']
```

```python
df_scaled_features.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_scaled_features
y = df['faction']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=47)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
K = 15
knn = KNeighborsClassifier(K)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
